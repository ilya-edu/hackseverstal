#### Процент отсутствующих значений

```python
# X - входные параметры
x_na_perc = df_x.isna().sum() / df_x.notnull().sum() * 100
pprint(f'Средний процент пропусков для X: {np.mean(x_na_perc.to_list()):.2f}')
pprint(x_na_perc)
```

```python
# Y - выходные параметры
y_na_perc = df_y.isna().sum() / df_y.notnull().sum() * 100
pprint(f'Средний процент пропусков для Y: {np.mean(y_na_perc.to_list()):.2f}')
pprint(y_na_perc)
```

#### Количество входных и выходных параметров

```python
x_cols = sorted(df_x.columns.to_list())
# print(*x_cols, sep='\n')
print(*sorted([[col.split('. ')[0], '. '.join(col.split('. ')[1:])] for col in x_cols]), sep='\n')
```

```python
x_cols_dict = defaultdict(list)
for k, v in sorted([[col.split('. ')[0], '. '.join(col.split('. ')[1:])] for col in x_cols]):
    x_cols_dict[k].append(v)
pprint(x_cols_dict)
```

```python
y_cols = sorted(df_y.columns.to_list())
# print(*y_cols, sep='\n')
print(*sorted([col.split('_')[1:] for col in y_cols]), sep='\n')
```

```python
y_cols_dict = defaultdict(list)
for k, v in sorted([col.split('_')[1:] for col in y_cols]):
    y_cols_dict[k].append(v)
pprint(y_cols_dict)
```

```python
# print('X:', *[f'Machine: {am[0]}, Cols: {len(am[1])}' for am in x_cols_dict.items()], sep='\n')
# print('Y:', *[f'Machine: {am[0]}, Cols: {len(am[1])}' for am in y_cols_dict.items()], sep='\n')

raw_x_data = [(am[0], len(am[1])) for am in x_cols_dict.items()]
raw_y_data = [(am[0].replace('А/М №', ''), len(am[1])) for am in y_cols_dict.items()]

x_data = pd.DataFrame(raw_x_data, columns=['Machine', 'X column count'])
y_data = pd.DataFrame(raw_y_data, columns=['Machine', 'Y column count'])
res =  pd.merge(x_data, y_data, left_on='Machine', right_on='Machine')
res
```
